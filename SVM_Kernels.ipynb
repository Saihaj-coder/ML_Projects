```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
```

```python
points_X, points_Y = make_circles(n_samples=1000, noise=0.1, random_state=15, factor=0.1)
```

```python
# points_Y
```

```python
plt.scatter(points_X[:, 0], points_X[:, 1], c = points_Y)
```

```python
points_X.shape
```

```python
X1 = points_X[:, 0].reshape(-1, 1)
X2 = points_X[:, 1].reshape(-1, 1)
print(X1.shape)
X3 = X1 + X2

print(X3.shape)
```

```python
points_X = np.hstack((points_X, X3))
```

```python
points_X.shape
```

```python
plt.scatter(points_X[:, 0], points_X[:, 2], c = points_Y)
```

```python
# Visualization
fig = plt.figure(figsize = (10,10))
axes = fig.add_subplot(111, projection = '3d')
axes.scatter(X1, X2, X3, c=points_Y, depthshade = True)
```

```python
from sklearn import svm

svc = svm.SVC(kernel = 'linear')
svc.fit(points_X, points_Y)
print(svc.score(points_X, points_Y))

w = svc.coef_
b = svc.intercept_
```

```python
w
```

```python
my_svm = svm.SVC(kernel='linear', C = 1, degree=5)
my_svm.fit(points_X, points_Y)
```

```python
my_svm.score(points_X, points_Y)
```

```python
my_svm = svm.SVC(kernel='poly', C = 1, degree = 10)
my_svm.fit(points_X, points_Y)
```

```python
my_svm.score(points_X, points_Y)
```

```python
my_svm = svm.SVC(kernel='rbf', C = 1, degree=5)
my_svm.fit(points_X, points_Y)
```

```python
my_svm.score(points_X, points_Y)
```

```python
#Plotting hyperplane

x1 = points_X[:,0].reshape((-1,1))
x2 = points_X[:,1].reshape((-1,1))
x1, x2 = np.meshgrid(x1, x2)
x3 = -(w[0][0]*x1 + w[0][1]*x2 + b)/w[0][2]

fig = plt.figure(figsize = (10,10))
axes2 = fig.add_subplot(111, projection = '3d')
axes2.scatter(X1, X2, X3, c=points_Y, depthshade = True)
axes1 = fig.add_subplot(111, projection = '3d')
axes1.scatter(x1, x2, x3, alpha = 0.5)
```
