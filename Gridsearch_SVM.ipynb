```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
```

```python
dataset = load_breast_cancer()
```

```python
dataset
```

```python
dataset.keys()
```

```python
print(dataset['feature_names'])
```

```python
data_X = pd.DataFrame(dataset['data'], columns=dataset['feature_names'])

data_Y = pd.DataFrame(dataset['target'], columns=['label'])
```

```python
data_X
```

```python
data_Y
```

```python
data_X = data_X.values
data_Y = data_Y.values

print(data_X.shape, data_Y.shape)
```

```python
train_X = data_X[:469, :]
train_Y = data_Y[:469, :]
test_X = data_X[469:, :]
test_Y = data_Y[469:, :]
```

```python
train_Y = train_Y.reshape(-1, )
test_Y = test_Y.reshape(-1, )
print(train_X.shape, test_X.shape, train_Y.shape, test_Y.shape)
```

```python
from sklearn import svm

my_svm = svm.SVC(kernel='rbf', C = 1)
my_svm.fit(train_X, train_Y)
```

```python
print(my_svm.score(train_X, train_Y))
```

```python
print(my_svm.score(test_X, test_Y))
```

```python
preds = my_svm.predict(test_X)
```

```python
preds, test_Y
```

```python
for i in range(len(preds)):
    print(preds[i], test_Y[i])
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(test_Y, preds))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_param = {
    'C':[0.1, 0.5, 1, 5],
    'kernel':['rbf', 'linear', 'poly', 'sigmoid'],
    'degree':[1,2,3,4,5,6]
}
```

```python
# grid_search = GridSearchCV?
grid_search = GridSearchCV(svm.SVC(), hyper_param, verbose=5)

grid_search.fit(train_X, train_Y)
```

```python
print(grid_search.best_params_)
```

```python
my_svm = svm.SVC(kernel='linear', C = 5, degree=1)
my_svm.fit(train_X, train_Y)
```

```python
print(my_svm.score(train_X, train_Y))
```

```python
print(my_svm.score(test_X, test_Y))
```

```python
preds = my_svm.predict(test_X)
```

```python
print(classification_report(test_Y, preds))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(test_Y, preds)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Reds', cbar=False,
            xticklabels=['Class 0', 'Class 1'], yticklabels=['Class 0', 'Class 1'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
